### Retention

```python
def calculate_retention(registration_file, auth_file, start_date=None, end_date=None):
    """
    Функция для расчета retention с возможностью указания диапазона дат.

    Аргументы:
    registration_file -- CSV файл с данными о регистрации пользователей.
    auth_file -- CSV файл с данными об авторизациях пользователей.
    start_date -- дата с которой начинаем считать retention (строка в формате 'YYYY-MM-DD').
    end_date -- дата до которой считаем retention (строка в формате 'YYYY-MM-DD').

    Возвращает:
    DataFrame с процентом вернувшихся пользователей по когортам за последние N месяцев.
    """

    # Чтение данных
    reg_data = pd.read_csv(registration_file, sep=';', encoding='ISO-8859-1')
    auth_data = pd.read_csv(auth_file, sep=';', encoding='ISO-8859-1')

    # Преобразование timestamp в даты
    reg_data['reg_date'] = pd.to_datetime(reg_data['reg_ts'], unit='s').dt.normalize()
    auth_data['auth_date'] = pd.to_datetime(auth_data['auth_ts'], unit='s').dt.normalize()

    # **Фильтрация по диапазону дат**
    if start_date:
        start_date = pd.to_datetime(start_date)
        reg_data = reg_data[reg_data['reg_date'] >= start_date]
        auth_data = auth_data[auth_data['auth_date'] >= start_date]

    if end_date:
        end_date = pd.to_datetime(end_date)
        reg_data = reg_data[reg_data['reg_date'] <= end_date]
        auth_data = auth_data[auth_data['auth_date'] <= end_date]

    # Объединение данных
    combined = pd.merge(auth_data, reg_data, on='uid', how='inner')

    # Вычисление дней с момента регистрации для каждой авторизации
    combined['delta_days'] = (combined['auth_date'] - combined['reg_date']).dt.days

    # Создание сводной таблицы для расчета retention
    pivot_table = combined.pivot_table(
        index='reg_date',
        columns='delta_days',
        values='uid',
        aggfunc='nunique'
    )

    # Количество новых пользователей в каждой когорте
    cohort_sizes = reg_data.groupby('reg_date')['uid'].nunique()

    # Делим каждое значение в строке на размер когорты, чтобы получить Retention Rate
    retention_rate = pivot_table.divide(cohort_sizes, axis=0)

    # Преобразование в проценты
    retention_rate = retention_rate * 100

    # **Визуализация** 
    fig, ax = plt.subplots(figsize=(20,20))
    sns.heatmap(retention_rate.iloc[:,1:], annot=True, cbar=False, square=True, cmap='Spectral', fmt='.1f')
    plt.title('Доля retention', size=20)
    plt.ylabel('Дата регистрации', size=10)
    plt.xlabel('Количество дней с момента регистрации', size=10)
    plt.show()

    # Возвращаем таблицу retention_rate
    return retention_rate
```

В качестве примера посчитаем retention за последний месяц. Самая поздняя дата регистрации у нас уже есть.
